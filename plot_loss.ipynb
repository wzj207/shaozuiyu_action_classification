```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# res50
csv_path = 'materials_shaozuiyu_exp/his_training_res50.csv'
df1 = pd.read_csv(csv_path)
df1.head()
```

```python
# res101
csv_path = 'materials_shaozuiyu_exp/his_training_res101.csv'
df2 = pd.read_csv(csv_path)
df2.head()
```

```python
# res152
csv_path = 'materials_shaozuiyu_exp/his_training_res152.csv'
df3 = pd.read_csv(csv_path)
df3.head()
```

```python
from matplotlib.font_manager import FontProperties


from pylab import mpl
mpl.rcParams['font.sans-serif'] = ['SimHei']

mpl.rcParams['axes.unicode_minus'] = False  


font = FontProperties()
# font.set_family('serif')
font.set_name('Times New Roman')
# font.set_style('italic')

font = {'family': 'Times New Roman',
#         'color':  'darkred',
        'color':  'black',
        'weight': 'normal',
        'size': 16,
        }

legend_font = {'family': 'Times New Roman',
        'weight': 'normal',
        'size': 12,
        }


lw = 1.5

f, ax = plt.subplots(nrows=2, ncols=1, figsize=(12,0.618*12))


ax[0].plot(df1['Train Loss'], color='orange', label='Training-ResNet50', linewidth=lw, )
ax[0].plot(df2['Train Loss'], color='deeppink', label='Training-ResNet101', linewidth=lw)
ax[0].plot(df3['Train Loss'], color='cyan', label='Training-ResNet152', linewidth=lw)
ax[0].set_xlim(0,100)
ax[0].set_ylim(0,1.0)

ax[0].plot(df1['Val. Loss'], color='orange', label='Validation-ResNet50', linewidth=lw, linestyle='dashed')
ax[0].plot(df2['Val. Loss'], color='deeppink', label='Validation-ResNet101', linewidth=lw, linestyle='dashed')
ax[0].plot(df3['Val. Loss'],  color='cyan', label='Validation-ResNet152', linewidth=lw, linestyle='dashed')
ax[0].set_xlim(0,100)
ax[0].set_ylabel(u'\nLoss', fontdict=font)
# ax[0].set_xticks()
ax[0].legend(loc='upper right', prop=legend_font)
ax[0].tick_params(axis='x',size=4)
ax[0].tick_params(axis='y',size=4)



# ax[1].plot(df1['Val. Loss'], color='black')
# ax[1].plot(df2['Val. Loss'], color='red')
# ax[1].plot(df3['Val. Loss'], )
# ax[1].set_xlim(0,100)


ax[1].plot(df1['Val. Acc'], color='orange', label='ResNet50', linewidth=lw)
ax[1].plot(df2['Val. Acc'], color='deeppink', label='ResNet101', linewidth=lw)
ax[1].plot(df3['Val. Acc'], color='cyan', label='ResNet152', linewidth=lw)
ax[1].set_xlim(0,100)
ax[1].set_ylabel(u'\n Validation Accuracy (%)', fontdict=font)
ax[1].set_xlabel(u' Epoch', fontdict=font)
ax[1].tick_params(axis='x',size=4)
ax[1].tick_params(axis='y',size=4)

ax[1].legend(loc='lower right',prop=legend_font)

plt.savefig('Fig_Training_Curve.tif')
```
